## 2. Formalización
|                                                                                                                                                                                                                                                                                                                                                                                                                                       **DESCRIPCIÓN DE LA SOLICITUD**                                                                                                                                                                                                                                                                                                                                                                                                                                      |
|:----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------:|
|                                                                                                                                                                                                                                                                                                                                                                                                                                           **Solcitante: Rodrigo Martínez**                                                                                                                                                                                                                                                                                                                                                                                                                                          |
| Es un despacho de abogados que quiere automatizar las demandas de sus clientes, esto lo harán a través de una página web, llenando un formulario. Al momento de llenar el formulario se manda al proceso de pago para finalizar la transacción. Para dar seguimiento a su demanda, el cliente crea una cuenta en la plataforma y verá el seguimiento de cada una de las actualizaciones del proceso legal. El administrador del sitio recibe la notificación de una nueva demanda y con los datos llenados del formulario se crea automáticamente el documento legal en formato word para empezar el proceso. El administrador recibe el pago y debe de ser capaz de verlo en un dashboard para ver la cantidad de ingresos recibidos. El administrador actualiza el proceso de la demanda y agrega comentarios en cada paso del proceso. La página debe ser responsive para poder verla desde el celular. |
|                                                                                                                                                                                                                                                                                                                                                                                                                                             **Líder funcional: Marco Pérez**                                                                                                                                                                                                                                                                                                                                                                                                                                            |
| Se requiere crear una aplicación capaz de automatizar la solicitud, proceso y recepción de una demanda desde la emisión y pago de los servicios por parte del solicitante, hasta la recepción por parte del asesor legal que prestará sus servicios, todo esto en formato word. La aplicación debe notificar al solicitante cada vez que haya una modificación en el proceso legal, en donde podrá conocer detalles adicionales proporcionados por el asesor, acerca de cada paso. Debe ser accesible especialmente desde dispositivos móviles y ser fácil de manipular.

### Casos

| **N°** |                     **Título**                     |  **Fecha** |                                                                                                                                                                           **Descripción**                                                                                                                                                                          |                                                                                                                                     **Objetivo**                                                                                                                                     |
|:------:|:--------------------------------------------------:|:----------:|:------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------:|:------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------:|
|  **1** | Necesidad de contacto directo con el asesor legal. | 25/02/2022 | Me he topado con muchos despachos que carecen de un vínculo ágil entre el abogado y el cliente y en momentos de urgencia no se puede tener contacto con ellos, por ejemplo en el caso de abogados penales cuando un acusado es tratado como culpable sin un veredicto y no sabes qué hacer porque cualquier cosa puede afectar al que creen que cometió el delito. | El uso del bot como filtro de las solicitudes ayuda a eliminar esta carga de manejo de información al personal humano, haciendo que la comunicación con el cliente pueda ser menos forzada y además que cuando se haga ya haya una idea de las necesidades del caso y de la persona. |
|  **2** |          Confusión en tarifas de servicio          | 25/02/2022 |                                                                    Algunos abogados carecen de un precio fijo cuando prestan sus servicios entonces a medida que un caso se complica y se alarga, la tarifa se modifica de forma drástica y en algunos casos hace más complicada la continuidad.                                                                   |                                               El bot podría ayudar a concentrar toda la información de precios según als necesidades que establezcas en tu solicitud, que posteriormente será analizada por un equipo dedicado a ello.                                               |
|  **3** |            Dudas en proceso de solicitud           | 25/02/2022 |                                                                       Uno como cliente muchas veces no conoce la variedad de abogados que hay en el mercado y por lo mismo en qué categoría entra el caso en el que necesitamos ayuda, esto hace complicada la búsqueda de asesores ideales.                                                                       |                                           El bot al mostrarte las opciones y descripciones de cada categoría, cuando procesa tu solicitud, haría mucho más simple comprender la necesidad que sabemos que tenemos pero no sabemos expresar.                                          |
